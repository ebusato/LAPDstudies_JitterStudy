```python
%matplotlib inline
import matplotlib.pyplot as plt

nTot = 300

T = 1/180000000.7 # sec (period of 180 MHz clock)
Tprime = 1/20 # sec (period of pulses - set on pulse generator)
#T = 0.5
#Tprime = 2.2

print("T = ", T, ", Tprime = ", Tprime, ", Tprime/T = ", Tprime/T)

deltat = []
id = []

for nprime in range(0, nTot):
    id.append(nprime)
    n = nprime*Tprime//T+1
    #print("nprime = ", nprime, ", n = ", n, ", n*T = ", n*T, ", nprime*Tprime = ", nprime*Tprime)
    deltat.append((n*T - nprime*Tprime)%T)

#print(id)
#print(deltat)

plt.plot(id, deltat)
```
